Perceptron Limitations:
- Perceptron is a linear model that can only learn linear decision boundaries.
- It struggles to accurately classify samples that are not linearly separable.
- This is evident in the XOR logical operation, where a linear boundary cannot separate the classes.
- More advanced techniques like neural networks with non-linear activation functions can handle such scenarios.
